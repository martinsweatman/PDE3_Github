__Q1__ (From exercise 9.4.3 of Advanced Modern Engineering Mathematics)

Consider the spherically symmetric form of the heat conduction equation
\begin{align}
\frac{\partial^2 u}{\partial r^2} + \frac{2}{r}\frac{\partial u}{\partial r} = \frac1\kappa\frac{\partial u}{\partial t} && (1)
\end{align}

a) Define $v(r,t) = r u(r,t)$ and show that $v$ satisfies the standard one-dimensional heat conduction equation. What can we expect of a solution as $r\to\infty$?
\
\
\
__Solution__

If $v=ru$, then differentiating produces

\begin{align}
\frac{\partial v}{\partial r} &= u + r\frac{\partial u}{\partial r} \\
\frac{\partial^2 v}{\partial r^2} &= 2 \frac{\partial u}{\partial r} + r\frac{\partial^2 u}{\partial r^2}
\end{align}

and hence using (1) we have

$$
\frac{\partial^2 u}{\partial r^2} + \frac{2}{r} \frac{\partial u}{\partial r}  = \frac{1}{r}\frac{\partial^2 v}{\partial r^2} = \frac{1}{\kappa r}\frac{\partial v}{\partial t}
$$

which after cancelling out the $r$ gives the standard one-dimensional heat conduction equation.

If $v$ remains bounded for $r\to\infty$ the solution $u\to0$ for $r\to\infty$.
\
\
\
b) Solve the equation in the annulus $a\le r\le b$ subject to the boundary conditions

\begin{align}
u(a,t) &= T_0, \quad & t>0 \\
u(b,t) &= 0, \quad & t>0 \\
u(r,0) &= 0, & a\le r\le b
\end{align}

Show that the solution has the form

$$
T(r,t) = \frac{a T_0}{r} \left[\frac{b-r}{b-a} - \sum_{N=1}^\infty A_N e^{-\kappa\lambda^2 t} \sin\left(\frac{r-a}{b-a}N\pi\right) \right]
$$

where $\lambda(b-a)=N\pi$. Evaluate the Fourier coefficients $A_N$.
\
\
\
__Solution__

We solve the standard one-dimensional heat conduction equation for $v(r,t) = r u(r,t)$. Since $u=\frac{v}{r}$, the boundary conditions for $v$ are

\begin{align}
u(a,t) &= T_0 \xrightarrow{yields} v(a,t) = aT_0, \quad & t>0 \\
u(b,t) &= 0 \xrightarrow{yields} v(b,t) = 0, \quad & t>0 \\
u(r,0) &= 0 \xrightarrow{yields} v(r,0) = 0, & a\le r\le b
\end{align}

First, we need to find the steady state solution $V(r)$ for which $v(r,t) = V(r) + w(r,t)$. Since the inner boundary is held at $aT_0$ and the outer at $0$, we get a linearly decreasing temperature from the inner to the outer boundary $V(r) = aT_0\frac{b - r}{b - a}$

With this steady state solution we get the following boundary conditions for the transient part $w(r,t)$

\begin{align}
w(a,t) &= 0, \quad & t>0 && (2)\\
w(b,t) &= 0, \quad & t>0 && (3)\\
w(r,0) &= -aT_0\frac{b - r}{b - a}, & a\le r\le b && (4)
\end{align}

From the lectures we know that the following three functions

\begin{align*}
w_1(r,t) =& (Ar+B) e^{\mu\kappa t}, & \mu = 0 && \\
w_2(r,t) =&\left(Ae^{\lambda r}+Be^{-\lambda r}\right) e^{\mu\kappa t}, & \mu=\lambda^2 \gt 0 && \\
w_3(r,t) =& (A\cos{(\lambda r)}+B\sin{(\lambda r)}) e^{\mu\kappa t}, & \mu=-\lambda^2 \lt 0 &&
\end{align*}

are possible solutions for the 1-d heat equation. We need to go through the boundary conditions to find which of these three functions is a solution to the particular problem and to find the values for the parameters, i.e. $A$, $B$ and $\mu$.

Clearly, $w_1$ is not a solution, since to satisfy both boundary conditions (2) and (3), we must have $A=B=0$. Also, since the solution is bounded, $w_2$ cannot be a solution either. Thus we are left with the only possibility,

$$
w(r, t) =\sum_{N=0}^\infty e^{-\kappa \lambda_N^2 t} \left(a_N\sin(\lambda_N (r + \phi)) + b_N\cos(\lambda_N r)\right)
$$

where we need to find the parameters $\lambda_N$, $a_N$ and $b_N$. We have added a shift of the sine so that we can better handle the boundary conditions. We can easily check that this is also a solution to the heat conduction equation.

We set $b_N=0$ for all $N$ since there is no cosine term in the sought solution and check if we can still match all boundary conditions. To match the condition at $r=a$ we shift the sine so that its argument is zero: $\phi = - a$. 

From the second boundary condition at $r=b$ we see that 

$$
\lambda_N = \frac{N\pi}{b-a}, \quad N=1,2,\dots
$$

Thus, we get

\begin{align}
w(r, t) =\sum_{N=1}^\infty a_N e^{-\kappa \left(\frac{N\pi}{b-a}\right)^2 t} \sin\left(\frac{r-a}{b-a}N\pi\right) && (5)
\end{align}

and the complete solution as

\begin{align}
u(r,t) = \frac{1}{r}(V(r) + w(r,t)) =  \frac{a T_0}{r} \left[\frac{b-r}{b-a} - \sum_{N=1}^\infty A_N e^{-\kappa \left(\frac{N\pi}{b-a}\right)^2 t} \sin\left(\frac{r-a}{b-a}N\pi\right) \right]
\end{align}

where $a_N = -aT_0 A_N$ for all $N$. (Note, if we repeat the above process for the cosine term, including a shift in its argument, we find we get exactly the same argument as for the sine term except it is shifted by $0.5\pi$. Thus, the cosine term is redundant, which is why it does not appear in the sought solution.)

To find the parameters $A_N$ we need to use (4). Setting $t=0$ in (5) generates the Fourier problem

\begin{align}
\frac{b-r}{b-a} &= \sum_{N=1}^\infty A_N \sin\left(\frac{r-a}{b-a}N\pi\right)
\end{align}

To solve this problem let's first transform it into standard form by setting $x=r-a$

\begin{align}
1-\frac{x}{b-a} &= \sum_{N=1}^\infty A_N \sin\left(\frac{N\pi x}{b-a} \right)
\end{align}

Given the presence of the sine function, we can perform an odd series half-range extension with half-period $b-a$ and obtain the Fourier coefficients

$$
A_N = \frac{2}{b-a}\int_0^{b-a}\left(1-\frac{x}{b-a}\right)\sin\left(\frac{N\pi x}{b-a}\right) dx
$$

Integrating by parts yields the coefficients

$$
A_N = \frac{2}{N\pi}, \quad N=1,2,3,\dots
$$

Thus, the full solution is

$$
u(r,t) =  \frac{a T_0}{r} \left[\frac{b-r}{b-a} - \sum_{N=1}^\infty \frac{2}{N\pi} e^{-\kappa\left(\frac{N\pi}{b-a}\right)^2 t} \sin\left(\frac{r-a}{b-a}N\pi\right) \right]
$$
\

c) Use Python to plot the solution.

__Solution__

Using the Python code from Exercise sheet 5 and choosing a suitbale upper limit for $t$ gives

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator , FormatStrFormatter
import numpy as np
# Solution to the heat conduction equation in a cyclindrical annulus
# Set the r and t grid points and the meshgrid
N = 100
a = 1
b = 2
r = np.linspace(a,b,N)
t = np.linspace(0,0.2,N)
R,T = np.meshgrid(r,t)
kappa = 0.3
T0 = 2
# Define the function and generate the output on the meshgrid
Z = R*0
M = 200
for i in range(M):
    p = (i+1)*np.pi/(b-a)
    Z = Z + 2/(i+1)/np.pi * np.exp(-kappa*p*p*T) * np.sin(p*(R-a))
Z = ((b-R)/(b-a)-Z)*a*T0/R
# Generate the plot
fig = plt.figure()
ax = plt.axes(projection='3d')
# Plot the surface
surf = ax.plot_surface(R,T,Z,cmap=cm.coolwarm,linewidth=0,antialiased=False)
# Customize the z axis
ax.set_zlim(0,2)
ax.zaxis.set_major_locator(LinearLocator(6))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.01f'))
ax.set_xlabel('r')
ax.set_ylabel('t')
ax.set_zlabel('u(r,t)')
ax.view_init(10,-75)
# Add a color bar which maps values to colors
fig.colorbar(surf,shrink=0.75,aspect=10)
plt.show()
```
