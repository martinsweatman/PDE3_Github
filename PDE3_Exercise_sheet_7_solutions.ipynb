__Q1__ (From exercise 9.4.3 of Advanced Modern Engineering Mathematics)

Consider the spherically symmetric form of the heat conduction equation
\begin{align}
\frac{\partial^2 u}{\partial r^2} + \frac{2}{r}\frac{\partial u}{\partial r} = \frac1\kappa\frac{\partial u}{\partial t} && (1)
\end{align}

a) Define $v(r,t) = r u(r,t)$ and show that $v$ satisfies the standard one-dimensional heat conduction equation. What can we expect of a solution as $r\to\infty$?
\
\
\
__Solution__

If $u=v/r$, then differentiating produces

\begin{align}
\frac{\partial u}{\partial r} &= -\frac{v}{r^2} + \frac{\partial v}{r\partial r} \\
\frac{\partial^2 u}{\partial r^2} &= 2\frac{v}{r^3} -2\frac{\partial v}{r^2\partial r} + \frac{\partial^2 v}{r\partial r^2}
\end{align}

and hence using (1) we have

\begin{align}
2\frac{v}{r^3} - 2\frac{\partial v}{r^2\partial r} + \frac{\partial^2 v}{r\partial r^2} + \frac{2}{r} \left( -\frac{v}{r^2} + \frac{\partial v}{r\partial r}  \right) &= \frac{1}{\kappa r}\frac{\partial v}{\partial t} \\
\frac{\partial^2 v}{r\partial r^2} = \frac{1}{\kappa r}\frac{\partial v}{\partial t}
\end{align}

which after cancelling out the $r$ gives the standard one-dimensional heat conduction equation.

If $v$ remains bounded for $r\to\infty$ the solution $u\to0$ for $r\to\infty$.
\
\
\
b) Solve the equation in the annulus $a\le r\le b$ subject to the boundary conditions

\begin{align}
u(a,t) &= T_0, \quad & t>0 \\
u(b,t) &= 0, \quad & t>0 \\
u(r,0) &= 0, & a\le r\le b
\end{align}

Show that the solution has the form

$$
u(r,t) = \frac{a T_0}{r} \left[\frac{b-r}{b-a} - \sum_{N=1}^\infty A_N e^{-\kappa\lambda^2 t} \sin\left(\frac{r-a}{b-a}N\pi\right) \right]
$$

where $\lambda(b-a)=N\pi$. Evaluate the Fourier coefficients $A_N$.
\
\
\
__Solution__

We solve the standard one-dimensional heat conduction equation for $v(r,t) = r u(r,t)$. Since $u=\frac{v}{r}$, the boundary conditions for $v$ are

\begin{align}
u(a,t) &= T_0 \xrightarrow{yields} v(a,t) = aT_0, \quad & t>0 \\
u(b,t) &= 0 \xrightarrow{yields} v(b,t) = 0, \quad & t>0 \\
u(r,0) &= 0 \xrightarrow{yields} v(r,0) = 0, & a\le r\le b
\end{align}

Since there are no time-varying inputs we can split the solution into steady-state and transient parts, $v(r,t) = V(r) + w(r,t)$, where the steady-state part, $V(r)$, is linear in $r$. Since the inner boundary is held at $aT_0$ and the outer at $0$, for the steady-state part we get a linearly decreasing temperature from the inner to the outer boundary, $V(r) = aT_0\frac{b - r}{b - a}$.

With this steady state solution we get the following boundary conditions for the transient part $w(r,t)$

\begin{align}
w(a,t) &= 0, \quad & t>0 && \\
w(b,t) &= 0, \quad & t>0 && \\
w(r,0) &= -aT_0\frac{b - r}{b - a}, & a\le r\le b &&
\end{align}

To simplify the algebra, it is convenient to use the substitution $x = r - a$. Then the inner boundary is at $x=0$ and the outer boundary is at $x=b-a$. Moreover, the PDE is transformed to,

\begin{align}
\left( \frac{\partial x}{\partial r} \right)^2 \frac{\partial^2 w(x,t)}{\partial x^2} = \frac{1}{\kappa}\frac{\partial w(x,t)}{\partial t}
\end{align}

Since $\left( \frac{\partial x}{\partial r} \right)^2 = 1$, this is just the usual 1-d heat conduction equation for $w(x,t)$. The boundary conditions are now,

\begin{align}
w(0,t) &= 0, \quad & t>0 && (2)\\
w(b-a,t) &= 0, \quad & t>0 && (3)\\
w(x,0) &= -aT_0 \left( 1 - \frac{x}{b - a} \right), & 0\le x\le b-a && (4)
\end{align}

From the lectures we know that the following three functions

\begin{align*}
w_0(x,t) =& (Ax+B) e^{\mu\kappa t}, & \mu = 0 && \\
w_+(x,t) =&\left(Ae^{\lambda x}+Be^{-\lambda x}\right) e^{\mu\kappa t}, & \mu=\lambda^2 \gt 0 && \\
w_-(x,t) =& (A\sin{(\lambda x)}+B\cos{(\lambda x)}) e^{\mu\kappa t}, & \mu=-\lambda^2 \lt 0 &&
\end{align*}

are possible solutions for the 1-d heat equation. We need to go through the boundary conditions to find which of these three functions is a solution to the particular problem and to find the values for the parameters, i.e. $A$, $B$ and $\mu$.

Clearly, $w_0$ is not a solution, since to satisfy both boundary conditions (2) and (3), we must have $A=B=0$ (in other words, by already finding the steady-state part, which is linear in $r$, we don't also need this linear solution; it is redundant). Also, since the solution is bounded, $w_+$ cannot be a solution either. Thus we are left with the only possibility,

$$
w(x, t) = \left(A\sin(\lambda x) + B\cos(\lambda x)\right) e^{-\lambda^2 \kappa t}
$$

where we need to find the parameters $\lambda$, $A$ and $B$. To match boundary condition (2) at $x=0$ we require $B=0$. Moreover, from boundary condition (3) at $x=b-a$ we see that 

$$
\lambda_N = \frac{N\pi}{b-a}, \quad N=1,2,\dots
$$

Thus, we obtain

\begin{align}
w(x, t) =\sum_{N=1}^\infty a_N \sin\left(\frac{N\pi x}{b-a}\right) e^{-\kappa \left(\frac{N\pi}{b-a}\right)^2 t} && (5)
\end{align}

and the complete solution is

\begin{align}
u(r,t) = \frac{1}{r}(V(r) + w(r,t)) =  \frac{a T_0}{r} \left[\frac{b-r}{b-a} - \sum_{N=1}^\infty A_N  \sin\left(\frac{r-a}{b-a}N\pi\right)e^{-\kappa \left(\frac{N\pi}{b-a}\right)^2 t} \right]
\end{align}

where $a_N = -aT_0 A_N$ for all $N$.

To find the parameters $A_N$ we need to use (4). Setting $t=0$ in (5) generates the Fourier problem

\begin{align}
-aT_0 \left( 1 - \frac{x}{b-a} \right) &= \sum_{N=1}^\infty a_N \sin\left(\frac{N\pi x}{b-a}\right)
\end{align}

Given the presence of the sine function, we can perform an odd series half-range extension with half-period $b-a$ and obtain the Fourier coefficients

$$
A_N = \frac{2}{b-a}\int_0^{b-a}\left(1-\frac{x}{b-a}\right)\sin\left(\frac{N\pi x}{b-a}\right) dx
$$

Integrating by parts yields the coefficients

$$
A_N = \frac{2}{N\pi}, \quad N=1,2,3,\dots
$$

Thus, the full solution is

$$
u(r,t) =  \frac{a T_0}{r} \left[\frac{b-r}{b-a} - \sum_{N=1}^\infty \frac{2}{N\pi}  \sin\left(\frac{r-a}{b-a}N\pi\right)e^{-\kappa\left(\frac{N\pi}{b-a}\right)^2 t} \right]
$$
\

c) Use Python to plot the solution.

__Solution__

Using the Python code from Exercise sheet 5 and choosing a suitbale upper limit for $t$ gives

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator , FormatStrFormatter
import numpy as np
# Solution to the heat conduction equation in a cyclindrical annulus
# Set the r and t grid points and the meshgrid
N = 100
a = 1
b = 2
r = np.linspace(a,b,N)
t = np.linspace(0,0.2,N)
R,T = np.meshgrid(r,t)
kappa = 0.3
T0 = 2
# Define the function and generate the output on the meshgrid
Z = R*0
M = 200
for i in range(M):
    p = (i+1)*np.pi/(b-a)
    Z = Z + 2/(i+1)/np.pi * np.exp(-kappa*p*p*T) * np.sin(p*(R-a))
Z = ((b-R)/(b-a)-Z)*a*T0/R
# Generate the plot
fig = plt.figure()
ax = plt.axes(projection='3d')
# Plot the surface
surf = ax.plot_surface(R,T,Z,cmap=cm.coolwarm,linewidth=0,antialiased=False)
# Customize the z axis
ax.set_zlim(0,2)
ax.zaxis.set_major_locator(LinearLocator(6))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.01f'))
ax.set_xlabel('r')
ax.set_ylabel('t')
ax.set_zlabel('u(r,t)')
ax.view_init(10,-75)
# Add a color bar which maps values to colors
fig.colorbar(surf,shrink=0.75,aspect=10)
plt.show()
```
