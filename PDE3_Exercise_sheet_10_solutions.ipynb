__Q1__

a) Consider the hyperbolic PDE

\begin{align}
A \frac{\partial^2 u}{\partial t^2} + 2B \frac{\partial^2 u}{\partial t \partial x} + C \frac{\partial^2u}{\partial x^2} = 0 && (1)
\end{align}

defined $\forall x$ and $t\geq0$ with the boundary and initial conditions

\begin{align}
\frac{\partial u(x,0)}{\partial x} &= F'(x), & \forall x && (2) \\
\frac{\partial u(x,0)}{\partial t} &= G(x), & \forall x && (3)
\end{align}

Show that 

\begin{align}
u(x,t)=\frac{1}{c_1 - c_2}\left( c_1 F(x+c_2t) - c_2 F(x+c_1t) + \int_{x+c_2t}^{x+c_1t} G(z) dz\right) && (4)
\end{align}

is a solution of (1 - 3) if the wave velocities are $c_{1,2}=\frac{-B \pm \sqrt{B^2-AC}}{A}$, where $B^2>AC$.

\
\
__Solution__

This question concerns a generalisation of D'Alembert's solution to the more general PDE (1). We are given the general solution (4), and asked to show that it satisfies the PDE (1) with  the BCs (2) and (3). We are told the PDE is hyperbolic, so the solution must consist of a pair of travelling waves.

Let's consider the BCs first. We can check BC (2) by taking the partial derivative w.r.to $x$ and then setting $t=0$. This gives

\begin{align}
\frac{\partial u(x,0)}{\partial x}=\frac{1}{c_1 - c_2}\left( c_1 F'(x) - c_2 F'(x) \right) = F'(x)
\end{align}

Clearly, BC (2) is satisfied. Now let's check BC (3) by taking the partial derivative w.r.to $t$ and then setting $t=0$. This gives

\begin{align}
\frac{\partial u(x,0)}{\partial t}=\frac{1}{c_1 - c_2}\left( c_1 c_2 F'(x) - c_1 c_2 F'(x) + c_1 G(x) - c_2 G(x) \right) = G(x)
\end{align}

Clearly, BC (3) is also satisfied.

Now let's check the solution (4) satisfies PDE (1). To do this we first need to find the 2nd-order differentials of $u(x,t)$ given by (4). First, let's find the 1st-order differentials (note that since the solution is a pair of traveling waves, we can use the notation $F_1 = F(x+c_1t)$ and $F_2 = F(x+c_2t)$ etc.),

\begin{align}
u_x=\frac{1}{c_1 - c_2}\left( c_1 F_2' - c_2 F_1' + G_1 - G_2 \right) \\
u_t=\frac{1}{c_1 - c_2}\left( c_1 c_2 F_2' - c_1 c_2 F_1' + c_1 G_1 - c_2 G_2 \right) \\
\end{align}

Therfore, the 2nd-order differentials are,

\begin{align}
u_{xx}=\frac{1}{c_1 - c_2}\left( c_1 F_2'' - c_2 F_1'' + G_1' - G_2' \right) \\
u_{xt}=\frac{1}{c_1 - c_2}\left( c_1 c_2 F_2'' - c_1 c_2 F_1'' + c_1 G_1' - c_2 G_2' \right) \\
u_{tt}=\frac{1}{c_1 - c_2}\left( c_1 c_2^2 F_2'' - c_1^2 c_2 F_1'' + c_1^2 G_1' - c_2^2 G_2' \right)
\end{align}

Inserting these into the left-hand-side of the PDE (1) gives,

\begin{align}
\frac{1}{c_1 - c_2}\left( 
C \left (c_1 F_2'' - c_2 F_1'' + G_1' - G_2' \right) +
2B \left( c_1 c_2 F_2'' - c_1 c_2 F_1'' + c_1 G_1' - c_2 G_2' \right) +
A \left( c_1 c_2^2 F_2'' - c_1^2 c_2 F_1'' + c_1^2 G_1' - c_2^2 G_2' \right)
\right) \overset{!}{=}0
\end{align}

We need to show this equality '$=0$' holds. Collecting terms gives,

\begin{align}
\left( 
F_2'' c_1 \left (C + 2B c_2 + A c_2^2 \right) - F_1'' c_2 \left (C + 2B c_1 + A c_1^2 \right) +
G_1' \left (C + 2B c_1 + A c_1^2 \right) - G_2' \left (C + 2B c_2 + A c_2^2 \right)
\right) \overset{!}{=}0
\end{align}

Thus, the equality holds if $C+2B c_{1,2} + A c_{1,2}^2 = 0$. Therfore, the solution (4) satisfies the PDE (1) only if $c_{1,2}=\frac{-B \pm \sqrt{B^2-AC}}{A}$, as required. Note, since we are told $B^2>AC$, both roots are real, i.e. the velocity of each wave is real which guarantees the PDE is hyperbolic.
